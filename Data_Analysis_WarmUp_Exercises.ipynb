**"Function to Calculate Half of Marks Using Floor Division"**

```python
def marks(x):
  return x//2
```

**"Creating a New Column 'Half_marks' by Applying the 'marks' Function to 'Marks' "**

```python
df['Half_marks']=df['Marks'].apply(marks)
```
